### Dataset Kelurahan/Desa

1. Lakukan pengelompokan data berdasarkan Kabupaten/Kota dan diambil tahun 2019-2020

```python
# Sebelum dikelompokan, diubah dahulu tipe tahun dan variabel jumlah pada dataset ini menjadi int64
bumil_periksa4x['tahun'] = bumil_periksa4x['tahun'].apply(np.int64)
bumil_periksa4x['jumlah_ibu_hamil_periksa_4_kali'] = bumil_periksa4x['jumlah_ibu_hamil_periksa_4_kali'].apply(np.int64)

bumil_kek['tahun'] = bumil_kek['tahun'].apply(np.int64)
bumil_kek['jumlah_ibu_hamil_mengalami_kek'] = bumil_kek['jumlah_ibu_hamil_mengalami_kek'].apply(np.int64)

bumil_kek_kunj['tahun'] = bumil_kek_kunj['tahun'].apply(np.int64)
bumil_kek_kunj['jumlah_ibu_hamil_kek_kunjungan'] = bumil_kek_kunj['jumlah_ibu_hamil_kek_kunjungan'].apply(np.int64)

bumil_resti['tahun'] = bumil_resti['tahun'].apply(np.int64)
bumil_resti['jumlah_ibu_hamil_resti'] = bumil_resti['jumlah_ibu_hamil_resti'].apply(np.int64)

bumil_resti_kunj['tahun'] = bumil_resti_kunj['tahun'].apply(np.int64)
bumil_resti_kunj['jumlah_ibu_hamil_resti_kunjungan'] = bumil_resti_kunj['jumlah_ibu_hamil_resti_kunjungan'].apply(np.int64)

bumil_pil_fe['tahun'] = bumil_pil_fe['tahun'].apply(np.int64)
bumil_pil_fe['jumlah_ibu_hamil_pil_fe'] = bumil_pil_fe['jumlah_ibu_hamil_pil_fe'].apply(np.int64)

bumil_kons['tahun'] = bumil_kons['tahun'].apply(np.int64)
bumil_kons['jumlah_ibu_hamil_konseling_gizi'] = bumil_kons['jumlah_ibu_hamil_konseling_gizi'].apply(np.int64)
```

```python
# lakukan grouping menjadi kabupaten/kota dan tahun
bumil_periksa4x_gr = bumil_periksa4x.groupby(['bps_nama_kabupaten_kota', 'tahun']).sum()
bumil_kek_gr = bumil_kek.groupby(['bps_nama_kabupaten_kota', 'tahun']).sum()
bumil_kek_kunj_gr = bumil_kek_kunj.groupby(['bps_nama_kabupaten_kota', 'tahun']).sum()
bumil_resti_gr = bumil_resti.groupby(['bps_nama_kabupaten_kota', 'tahun']).sum()
bumil_resti_kunj_gr = bumil_resti_kunj.groupby(['bps_nama_kabupaten_kota', 'tahun']).sum()
bumil_pil_fe_gr = bumil_pil_fe.groupby(['bps_nama_kabupaten_kota', 'tahun']).sum()
bumil_kons_gr = bumil_kons.groupby(['bps_nama_kabupaten_kota', 'tahun']).sum()

# lakukan reset index
bumil_periksa4x_gr.reset_index(inplace=True)
bumil_kek_gr.reset_index(inplace=True)
bumil_kek_kunj_gr.reset_index(inplace=True)
bumil_resti_gr.reset_index(inplace=True)
bumil_resti_kunj_gr.reset_index(inplace=True)
bumil_pil_fe_gr.reset_index(inplace=True)
bumil_kons_gr.reset_index(inplace=True)

# pilih data tahun 2019 dan 2020
bumil_periksa_19_20 = bumil_periksa4x_gr[(bumil_periksa4x_gr['tahun'] == 2019) | (bumil_periksa4x_gr['tahun'] == 2020)]
bumil_kek_19_20 = bumil_kek_gr[(bumil_kek_gr['tahun'] == 2019) | (bumil_kek_gr['tahun'] == 2020)]
bumil_kek_kunj_19_20 = bumil_kek_kunj_gr[(bumil_kek_kunj_gr['tahun'] == 2019) | (bumil_kek_kunj_gr['tahun'] == 2020)]
bumil_resti_19_20 = bumil_resti_gr[(bumil_resti_gr['tahun'] == 2019) | (bumil_resti_gr['tahun'] == 2020)]
bumil_resti_kunj_19_20 = bumil_resti_kunj_gr[(bumil_resti_kunj_gr['tahun'] == 2019) | (bumil_resti_kunj_gr['tahun'] == 2020)]
bumil_pil_fe_19_20 = bumil_pil_fe_gr[(bumil_pil_fe_gr['tahun'] == 2019) | (bumil_pil_fe_gr['tahun'] == 2020)]
bumil_kons_19_20 = bumil_kons_gr[(bumil_kons_gr['tahun'] == 2019) | (bumil_kons_gr['tahun'] == 2020)]

# drop columns yang tidak diperlukan
bumil_periksa_19_20.drop(columns=['id','kode_provinsi', 'bps_kode_kabupaten_kota', 'bps_kode_kecamatan', 'bps_kode_desa_kelurahan'],inplace=True) 
bumil_kek_19_20.drop(columns=['id','kode_provinsi', 'bps_kode_kabupaten_kota', 'bps_kode_kecamatan', 'bps_kode_desa_kelurahan'],inplace=True)
bumil_kek_kunj_19_20.drop(columns=['id','kode_provinsi', 'bps_kode_kabupaten_kota', 'bps_kode_kecamatan', 'bps_kode_desa_kelurahan'],inplace=True)
bumil_resti_19_20.drop(columns=['id','kode_provinsi', 'bps_kode_kabupaten_kota', 'bps_kode_kecamatan', 'bps_kode_desa_kelurahan'],inplace=True)
bumil_resti_kunj_19_20.drop(columns=['id','kode_provinsi', 'bps_kode_kabupaten_kota', 'bps_kode_kecamatan', 'bps_kode_desa_kelurahan'],inplace=True)
bumil_pil_fe_19_20.drop(columns=['id','kode_provinsi', 'bps_kode_kabupaten_kota', 'bps_kode_kecamatan', 'bps_kode_desa_kelurahan'],inplace=True)
bumil_kons_19_20.drop(columns=['id','kode_provinsi', 'bps_kode_kabupaten_kota', 'bps_kode_kecamatan', 'bps_kode_desa_kelurahan'],inplace=True)

# Hilangkan kata "KABUPATEN"
bumil_periksa_19_20['bps_nama_kabupaten_kota'] = bumil_periksa_19_20['bps_nama_kabupaten_kota'].str.replace('KABUPATEN ', '')
bumil_kek_19_20['bps_nama_kabupaten_kota'] = bumil_kek_19_20['bps_nama_kabupaten_kota'].str.replace('KABUPATEN ', '')
bumil_kek_kunj_19_20['bps_nama_kabupaten_kota'] = bumil_kek_kunj_19_20['bps_nama_kabupaten_kota'].str.replace('KABUPATEN ', '')
bumil_resti_19_20['bps_nama_kabupaten_kota'] = bumil_resti_19_20['bps_nama_kabupaten_kota'].str.replace('KABUPATEN ', '')
bumil_resti_kunj_19_20['bps_nama_kabupaten_kota'] = bumil_resti_kunj_19_20['bps_nama_kabupaten_kota'].str.replace('KABUPATEN ', '')
bumil_pil_fe_19_20['bps_nama_kabupaten_kota'] = bumil_pil_fe_19_20['bps_nama_kabupaten_kota'].str.replace('KABUPATEN ', '')
bumil_kons_19_20['bps_nama_kabupaten_kota'] = bumil_kons_19_20['bps_nama_kabupaten_kota'].str.replace('KABUPATEN ', '')

```

```python
# Gabungkan ke 7 dataset menjadi satu dataset kelurahan/desa

kel_desa_1 = pd.merge(bumil_periksa_19_20, bumil_kek_19_20, how='inner', 
                      on=['bps_nama_kabupaten_kota','tahun'])
kel_desa_2 = pd.merge(bumil_kek_kunj_19_20, bumil_resti_19_20, how='inner', 
                      on=['bps_nama_kabupaten_kota','tahun'])
kel_desa_3 = pd.merge(bumil_resti_kunj_19_20, bumil_pil_fe_19_20, how='inner', 
                      on=['bps_nama_kabupaten_kota','tahun'])
kel_desa_4 = pd.merge(kel_desa_1, bumil_kons_19_20, how='inner', 
                      on=['bps_nama_kabupaten_kota','tahun'])
kel_desa_23 = pd.merge(kel_desa_2,kel_desa_3, how='inner',
                      on=['bps_nama_kabupaten_kota','tahun'])

kel_desa = pd.merge(kel_desa_4,kel_desa_23, how='inner',
                      on=['bps_nama_kabupaten_kota','tahun'])

# ubah nama kolom bps_nama_kabupaten_kota menjadi nama_kabupaten_kota
kel_desa.rename(columns={'bps_nama_kabupaten_kota': 'nama_kabupaten_kota'},inplace=True)
```
